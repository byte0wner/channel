```python
keyboard_layout = [
    ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0', '-', '='],
    ['q', 'w', 'e', 'r', 't', 'y', 'u', 'i', 'o', 'p', '[', ']', '\\'],
    ['a', 's', 'd', 'f', 'g', 'h', 'j', 'k', 'l', ';', "'"],
    ['z', 'x', 'c', 'v', 'b', 'n', 'm', ',', '.', '/']
]
```

```python
def read_config():
    config = []
    with open("config.kw", "r") as file:
        for line in file:
            clean_line = line.strip().split(":")
            config.append(clean_line)
        return config

paths = read_config()
print(paths)
```

```python
password = ""

# start coordinates
ptr_row = 0
ptr_col = 0

is_first_step = True

password = keyboard_layout[ptr_row][ptr_col]
for path in paths:
    steps = int(path[0])
    route = path[1]
    for step in range(0, steps):
        if is_first_step:
            is_first_step = False
            continue
        if route == "RIGHT":
            ptr_col = ptr_col + 1
        if route == "DOWN":
            ptr_row = ptr_row + 1
        if route == "LEFT":
            ptr_col = ptr_col - 1
        if route == "UP":
            ptr_row = ptr_row - 1
        password += keyboard_layout[ptr_row][ptr_col]
print(password)
```
